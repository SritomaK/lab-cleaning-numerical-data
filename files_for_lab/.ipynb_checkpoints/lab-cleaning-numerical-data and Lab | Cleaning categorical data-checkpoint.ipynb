### 4.Rename the columns so they follow the PE8 (snake case) (lower_case_with_underscores)



```python
#renaming the folumns using a function
#Snake Case

def snake_case(df):
    cols = []
    for i in range(len(df.columns)):
        cols.append(df.columns[i].lower().replace(' ', '_'))
    df.columns = cols
    return df

data = snake_case(data)
data
```

```python
# function for 'continuous_df' and 'discrete_df' with the separated data.

def continuous_discrete(df):
    continuous = []
    discrete = []

    for column in df.columns:
        unique_values = df[column].nunique()
        if unique_values > 10:
            continuous.append(column)
        else:
            discrete.append(column)

    continuous_df = data[continuous]
    discrete_df = data[discrete]

    return continuous_df, discrete_df


continuous_df, discrete_df = continuous_discrete(numerical_df)
```

### 12.Did you find outliers? Comment what you will do with them. If you decide to remove them, do that now. Also, remove them from your main dataframe.



```python
#'customer_lifetime_value', 'monthly_premium_auto', 'total_claim_amount': we have lots of outliers. removing the outliers
```

```python
def remove_outliers_quantile(df,columns):
    for column in columns:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound= Q1 - 1.5* IQR
        upper_bound = Q3 +1.5 * IQR
        df_clean=df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_clean

continuous_df = remove_outliers_quantile(continuous_df, ['customer_lifetime_value', 'monthly_premium_auto'])
```

```python
for column in continuous_df.columns:
    sns.histplot(continuous_df[column], kde=True)
    plt.show()
```
